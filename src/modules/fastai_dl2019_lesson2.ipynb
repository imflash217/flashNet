```python
from fastai import *
from fastai.vision import *
import numpy as np
```

```python
bs = 16
```

```python
# creating repositories and uploading URLs into the server
folder_black = "black"
folder_teddy = "teddy"
folder_grizzly = "grizzly"
file_black = "urls_black.csv"
file_teddy = "urls_teddy.csv"
file_grizzly = "urls_grizzly.csv"
```

```python
## Downloading images from Google using the CSV files
## We can use fastai's `download_images()` to do this

path = Path("../../../../data/bears")
for folder, file in [(folder_black, file_black), (folder_teddy, file_teddy), (folder_grizzly, file_grizzly)]:
    dest = path/folder
    dest.mkdir(parents=True, exist_ok=True)
    download_images(path/file, dest, max_pics=100)
    print(folder)
    verify_images(path/folder, delete=True)
```

```python
## creating our dataBunch & viewing our data

np.random.seed(217)
data = ImageDataBunch.from_folder(path=path, train=path, valid_pct=0.2, ds_tfms=get_transforms(),
                                  size=224, bs=bs).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(12,12))
```

```python
## Training resnet model 
```

```python
learn = cnn_learner(data=data, base_arch=models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(cyc_len=4)
```

```python
learn.save("bears_resnet34_stage1")
```

```python
learn.unfreeze()
```

```python
learn.lr_find(start_lr=1e-5, end_lr=1e-1)
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2)
```

```python
learn.load("bears_resnet34_stage1")
```

```python
interp = ClassificationInterpretation.from_learner(learn=learn)
```

```python
losses, idxs = interp.top_losses()
```

```python
len(data.valid_ds) == len(losses) == len(idxs)
```

```python
interp.plot_confusion_matrix(figsize=(5,5), dpi=60)
```

```python
interp.plot_top_losses(k=9, figsize=(12,12))
```
