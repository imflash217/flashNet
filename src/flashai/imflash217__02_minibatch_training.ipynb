## 2. Cross Entropy Loss:

### 2.1 Log-Softmax

````python
### Step-2.1: Log-Softmax loss
def log_softmax_v1(x):
    return (x.exp()/(x.exp().sum(dim=-1, keepdim=True))).log()

def log_softmax_v2(preds):
    ### using the trick: log(a/b) = log(a)-log(b)
    ### also, log(exp(x)) = x
    return preds - (preds.exp().sum(dim=-1, keepdim=True)).log()

def logsumexp(x):
    ### uses trick to calculate log(sum(exp(x))); which cares for the numerical stability in calculations
    ### log(sum(exp(x))) = m + log(sum(exp(x-m)))  where m = max(x)
    m = x.max(-1)[0]
    ### print(m.shape, m[0])
    return m + (x-m[:,None]).exp().sum(dim=-1).log()

def log_softmax(preds):
    ### using the logsumexp() trick to provide numerical stability
    ### print(logsumexp(preds)[:, None].shape)
    ### return preds - logsumexp(preds)[:, None]
    return preds - preds.logsumexp(dim=-1, keepdim=True)      ### using Pytorch's logsumexp() method after testing that our method also performs accurately
````

## 3. Basic Training Loop:
```
Basically the training loop iterates over the following steps:
1. Get the output of the model on a batch of inputs
2. Compare the output to the labels & compute the LOSS
3. Calculate the GRADIENTS of the loss wrt every parameter in the model
4. UPDATE the parmaeters with the gradients using a LEARNING RATE
5. Repeat 1-4
```

````python
loss_func = F.cross_entropy
````

````python
#export
def accuracy(preds, labels):
    return (torch.argmax(preds, dim=1) == labels).float().mean()
````

````python
bs = 16                                    ### batch-size
xb = x_train[0:bs]                         ### grabbing a batch from input data
yb = y_valid[0:bs]                         ### the corresponding labels of the input batch
preds = model(x=xb)                        ### calculating the predictions from model
loss = loss_func(input=preds, target=yb)   ### calculating the loss
acc = accuracy(preds=preds, labels=yb)     ### calculating the accuracy of preds from model for this batch
````

````python
lr = 0.5                                   ### learning rate
num_epochs = 1                             ### total number of epochs
````

````python
### Training loop
for epoch in range(num_epochs):
    for i in range((n-1) // bs + 1):
        start_i = i * bs
        end_i   = start_i + bs
        xb = x_train[start_i:end_i]                      ### input batch
        yb = y_train[start_i:end_i]                      ### label batch
        preds = model(xb)                                ### predictions from model
        loss = loss_func(preds, yb)                      ### batch loss
        acc  = accuracy(preds=preds, labels=yb)          ### batch accuracy
        
        loss.backward()                                  ### backward pass
        with torch.no_grad():
            for l in model.layers:
                if hasattr(l, "weight"):
                    l.weight -= l.weight.grad * lr       ### updating our weights
                    l.bias   -= l.bias.grad   * lr       ### updating our bias
                    l.weight.grad.zero_()                ### zeroing the weight gradients
                    l.bias.grad.zero_()                  ### zeroing the bias gradients
        
        ### print(f"loss = {loss} \t accuracy = {acc}")
        
````

### 4.1 `parameters`

````python
class Model(torch.nn.Module):
    def __init__(self, n_in, nh, n_out):
        super().__init__()
        self.l1 = torch.nn.Linear(in_features=n_in, out_features=nh)
        self.l2 = torch.nn.Linear(in_features=nh,   out_features=n_out)
        
    def __call__(self, x):
        return self.l2(F.relu(self.l1(x)))
````

````python
def fit():
    for epoch in range(num_epochs):
        for i in range((n-1)//bs + 1):
            start_i = i * bs
            end_i   = start_i + bs
            xb = x_train[start_i:end_i]
            yb = y_train[start_i:end_i]
            preds = model(x=xb)
            loss  = loss_func(input=preds, target=yb)
            acc   = accuracy(preds=preds, labels=yb)
            
            loss.backward()
            with torch.no_grad():
                for p in model.parameters():
                    p -= p.grad * lr
                model.zero_grad()
            print(f"loss = {loss} \t accuracy = {acc}")
````

```
Behind the scenes, Pytorch overrides the __setattr__ function in torch.nn.Module
so that the submodules defined by the user can be properly registered as parameters of the model
```

````python
class DummyModule():
    def __init__(self, n_in, nh, n_out):
        self._modules = {}
        self.l1 = torch.nn.Linear(in_features=n_in, out_features=nh)
        self.l2 = torch.nn.Linear(in_features=nh,   out_features=n_out)
        
    def __setattr__(self, key, value):
        if not key.startswith("_"):
            self._modules[key] = value
        super().__setattr__(key, value)
        
    def __repr__(self):
        return f"{self._modules}"
    
    def parameters(self):
        for l in self._modules.values():
            for p in l.parameters():
                yield p
````

````python
class Model(torch.nn.Module):
    def __init__(self, layers):
        super().__init__()
        self.layers = layers
        for i, l in enumerate(self.layers):
            self.add_module(f"layer_{i}", l)
    
    def __call__(self, x):
        for l in self.layers:
            x = l(x)
        return x
            
````

### 4.2 Optimizer `optim`

````python
class Optimizer():
    def __init__(self, params, lr=1e-3):
        self.params = list(params)            ### DO NOT forget to list()
        self.lr = lr
        
    def step(self):
        with torch.no_grad():
            for p in self.params:
                p -= p.grad * lr
    
    def zero_grad(self):
        for p in self.params:
            p.grad.data.zero_()
        
````

````python
optim = Optimizer(params=model.parameters(), lr=0.5)
````

## 5.1 `Dataset` class

It's clunky to iterate through minibatches of x and y separately:

```python
xb = x_train[strat_i:end_i]
yb = y_train[start_i:end_i]
```

So, we will try to build a `Dataset` class that does this for us like:

```python
xb, yb = train_ds[start_i:end_i]
```

````python
#export
class Dataset():
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
    
````

## 5.2 `Dataloader` class

Previously our training loop iterated over minibatch of data like this:

```python
for i in range((n-1)//bs + 1):
    xb, yb = train_ds[i*bs : (i+1)*bs]
    ...
```
We will replace this clunky piece by an elegant line below using a Dataloader class:

```python
for xb, yb in train_dl:
    ...
```

````python
#export
class Dataloader():
    def __init__(self, ds, bs):
        self.ds = ds
        self.bs = bs
        
    def __iter__(self):
        for i in range(0, len(self.ds), self.bs):
            yield self.ds[i : i+self.bs]
            
````

# 6. Random Sampling

We want our training set to be in a random order and that order should differ each iteration.

But the validation set shouldn't be randomized!!!

````python
class Sampler():
    def __init__(self, ds, bs, shuffle=False):
        self.n = len(ds)
        self.bs = bs
        self.shuffle = shuffle
        
    def __iter__(self):
        self.idxs = torch.randperm(self.n) if self.shuffle else torch.arange(self.n)
        for i in range(0, self.n, self.bs):
            yield self.idxs[i : i+self.bs]
````

````python
def collate(batch):
    xs, ys = zip(*batch)
    return torch.stack(xs), torch.stack(ys)

class Dataloader():
    def __init__(self, ds, sampler, collate_fn=collate):
        self.ds = ds
        self.sampler = sampler
        self.collate_fn = collate_fn
    
    def __iter__(self):
        for s in self.sampler:
            yield self.collate_fn([self.ds[i] for i in s])
````

## 6. `Validation`

````python
def fit(num_epochs, model, loss_fn, opt, train_dl, valid_dl):
    for epoch in range(num_epochs):
        loss_train = 0.
        acc_train  = 0.
        loss_valid = 0.
        acc_valid  = 0.
        n_train = len(train_dl)
        n_valid = len(valid_dl)
        
        ### Handle BatchNorm, Dropout etc
        ### Training
        model.train()
        for xb, yb in train_dl:
            preds = model(input=xb)
            loss  = loss_fn(input=preds, target=yb)
            acc   = accuracy(preds=preds, labels=yb)
            loss.backward()
            opt.step()
            opt.zero_grad()
            
            ###--------------------------###
            loss_train += loss.item()
            acc_train  += acc
            ###--------------------------###
        
        ### Validation
        model.eval()
        with torch.no_grad():
            for xb, yb in valid_dl:
                preds = model(input=xb)
                loss  = loss_fn(input=preds, target=yb)
                acc   = accuracy(preds=preds, labels=yb)
                
                ###--------------------------###
                loss_valid += loss.item()
                acc_valid  += acc
                ###--------------------------###
        
        ### Metrics Calculation
        loss_train /= n_train
        acc_train  /= n_train
        loss_valid /= n_valid
        acc_valid  /= n_valid
        
        print(f"loss_train={loss_train:.5f} \t loss_valid={loss_valid:.5f} \t \
                acc_train={acc_train:.5f} \t acc_valid={acc_valid:.5f}")            
    return loss_train, loss_valid, acc_train, acc_valid
````

````python
#export
def get_dls(train_ds, valid_ds, bs, **kwargs):
    return (DataLoader(dataset=train_ds, batch_size=bs, shuffle=True, **kwargs),
            DataLoader(dataset=valid_ds, batch_size=bs*2, shuffle=False, **kwargs))
````
